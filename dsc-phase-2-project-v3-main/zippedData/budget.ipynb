```python
# Imports

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create df
budget = pd.read_csv('tn.movie_budgets.csv.gz')
```

```python
budget.head()
```

```python
budget.info()
```

```python
#remove commas from numerical $ values
budget = budget.replace(',', '', regex=True)
```

```python
#remove $ from numerical values
budget['production_budget'] = budget['production_budget'].str[1:]
budget['domestic_gross'] = budget['production_budget'].str[1:]
budget['worldwide_gross'] = budget['production_budget'].str[1:]
```

```python
#convert the values in these columns from objects to integers
budget['production_budget'] = pd.to_numeric(budget['production_budget'])
budget['domestic_gross'] = pd.to_numeric(budget['domestic_gross'])
budget['worldwide_gross'] = pd.to_numeric(budget['worldwide_gross'])
```

```python
budget.info()
```

```python
budget.head()
```

```python
budget['return_on_investment'] = budget['worldwide_gross']/budget['production_budget']
```

```python
budget['release_date'] = pd.to_datetime(budget['release_date'])
```

```python
budget['year'] = budget['release_date'].dt.year
```

```python
budget.head()
```

```python
budget_recent = budget.loc[budget['year'] > 2010]
```

```python
budget_recent.head()
```

```python
budget_recent.sort_values(by='return_on_investment', ascending=False)
```
